 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and 
days.

 np.random.seed(20)â€¨
 data = {â€¨
 'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),â€¨
 'Day': np.random.choice(range(1, 31), 100),â€¨
 'Sales': np.random.randint(1000, 5000, 100)â€¨
 }
 â€¨
 df = pd.DataFrame(data)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate data
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df = pd.DataFrame(data)

# Pivot table for heatmap
sales_pivot = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(sales_pivot, cmap="coolwarm", annot=True, fmt=".0f")

plt.xlabel("Month")
plt.ylabel("Day")
plt.title("Sales Heatmap")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}

df = pd.DataFrame(data)

# 3D Surface plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("3D Surface Plot")

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate data
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}

df = pd.DataFrame(data)

# Bubble chart
plt.figure(figsize=(8, 5))
plt.scatter(df['GDP'], df['Population'], s=df['Population']*5, alpha=0.6, color='blue')

# Labeling
for i, txt in enumerate(df['Country']):
    plt.text(df['GDP'][i], df['Population'][i], txt, fontsize=12, ha='right')

plt.xlabel("GDP (in billion $)")
plt.ylabel("Population (in million)")
plt.title("Bubble Chart of Population vs GDP")
plt.grid(True)
plt.show()
```

 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Display the plot in the notebook
output_notebook()
show(p)


```

 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the 
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256

# Generate random x and y values
np.random.seed(42)
n = 100
x = np.random.rand(n) * 10
y = np.random.rand(n) * 10
sizes = np.random.rand(n) * 20 + 5  # Marker sizes
colors = np.random.choice(Viridis256, n)  # Random colors from palette

# Create a ColumnDataSource
data_source = ColumnDataSource(data={"x": x, "y": y, "sizes": sizes, "colors": colors})

# Create a Bokeh figure
p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y')
p.scatter("x", "y", source=data_source, size="sizes", color="colors", fill_alpha=0.6)

# Display the plot in the notebook
output_notebook()
show(p)

```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

 fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']â€¨

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Define dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
data_source = ColumnDataSource(data={"fruits": fruits, "counts": counts})

# Create a Bokeh figure
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', tools="")
p.vbar(x='fruits', top='counts', width=0.5, source=data_source, color="blue")

# Display the plot in the notebook
output_notebook()
show(p)

```

 4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate random data for histogram
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a Bokeh figure
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="blue", line_color="black")

# Display the plot in the notebook
output_notebook()
show(p)
```

 5. Create a Bokeh heatmap using the provided dataset.
 
 data_heatmap = np.random.rand(10, 10)â€¨
 x = np.linspace(0, 1, 10)â€¨
 y = np.linspace(0, 1, 10)â€¨
 xx, yy = np.meshgrid(x, y)

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, LinearColorMapper, ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Generate heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the arrays
x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()

# Create a ColumnDataSource
data_source = ColumnDataSource(data={"x": x_flat, "y": y_flat, "values": values})

# Define color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=min(values), high=max(values))

# Create a Bokeh figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', tools="", x_range=(0,1), y_range=(0,1))
p.rect(x='x', y='y', width=0.1, height=0.1, source=data_source, fill_color=linear_cmap('values', Viridis256, min(values), max(values)), line_color=None)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Display the plot in the notebook
output_notebook()
show(p)

```
